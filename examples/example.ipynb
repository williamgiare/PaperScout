# PaperScout Notebook Example

This notebook shows the notebook-friendly PaperScout API.

It covers the full workflow:

1. preview a query
2. estimate the number of matching papers
3. save matching results to a BibTeX file
4. generate a LaTeX `\\cite{...}` command from the saved file

## Setup

Install the package from the project root before running the notebook:

```bash
python3 -m pip install -e .
```

````python
from paperscout.api import estimate, preview, save
from paperscout.cite import cite
````

## 1. Preview a Query

The `preview(...)` wrapper is local only. It builds the normalized INSPIRE query and preview URLs without downloading results.

````python
prepared = preview(
    keywords=["inflation", "model"],
    author=["Starobinsky"],
    from_year=2022,
    min_citations=10,
)

print(prepared.human_query)
print()
print(prepared.preview_api_url)
print()
print(prepared.preview_bibtex_url)
````

## 2. Estimate Matching Papers

The `estimate(...)` wrapper performs the INSPIRE preflight request and returns the execution plan without downloading BibTeX.

````python
estimate_result = estimate(
    keywords=["inflation", "model"],
    author=["Starobinsky"],
    from_year=2022,
    min_citations=10,
)

estimate_result.execution_plan
````

## 3. Save Matching Results

The `save(...)` wrapper runs the full workflow and writes the BibTeX file to disk.

````python
save_result = save(
    output="output/starobinsky_inflation.bib",
    keywords=["inflation", "model"],
    author=["Starobinsky"],
    from_year=2022,
    min_citations=10,
)

save_result.saved_path
````

## 4. Build a LaTeX Citation Command

Once the `.bib` file exists, use `cite(...)` to generate a ready-to-use LaTeX citation command.

````python
cite(save_result.saved_path, sort="date")
````
